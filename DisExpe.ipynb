```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x=np.linspace(-100,100,2000)
mu1=0.0
sigma1=5.0
gaus1=1.0/np.sqrt(2.0*np.pi*sigma1*sigma1)*np.exp(-((x-mu1)*(x-mu1))/(2.0*sigma1*sigma1))
```

```python
plt.plot(x,gaus1)
```
